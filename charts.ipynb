```python
import pandas as pd
import plotly.graph_objects as go

# ------------------------
# Data
# ------------------------
data = [
    (2010,   63999,    305041),
    (2011,  618287,   3289387),
    (2012,  829857,   6978360),
    (2013, 2361863,   8821297),
    (2014, 5106741,   7455716),
    (2015, 6529649,  10009077),
    (2016, 4801429,   5007981),
    (2017, 6643079,  28547310),
    (2018, 6244716,  11191339),
    (2019, 9741005,  14001038),
    (2020,15919518,  29896630),
    (2021,22248656,  32529472),
    (2022,25782900,  58282649),
    (2023,20356069,  46737779),
]

df = pd.DataFrame(data, columns=["year", "credits_retired", "credits_issued"])
df["retired_share"] = df["credits_retired"] / df["credits_issued"]

# ------------------------
# Figure: Issued + Retired (lines) + Retired share (secondary axis)
# ------------------------
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df["year"],
        y=df["credits_issued"],
        mode="lines+markers",
        name="Credits Issued",
        customdata=df[["credits_retired", "retired_share"]].to_numpy(),
        hovertemplate=(
            "<b>%{x}</b><br>"
            "Issued: %{y:,}<br>"
            "Retired: %{customdata[0]:,}<br>"
            "Retired share: %{customdata[1]:.1%}"
            "<extra></extra>"
        ),
    )
)

fig.add_trace(
    go.Scatter(
        x=df["year"],
        y=df["credits_retired"],
        mode="lines+markers",
        name="Credits Retired",
        hovertemplate="<b>%{x}</b><br>Retired: %{y:,}<extra></extra>",
    )
)

fig.add_trace(
    go.Scatter(
        x=df["year"],
        y=df["retired_share"],
        mode="lines",
        name="Retired share (Retired / Issued)",
        yaxis="y2",
        hovertemplate="<b>%{x}</b><br>Retired share: %{y:.1%}<extra></extra>",
    )
)

fig.update_layout(
    title="Carbon Credits Issued vs Retired (2010–2023)",
    xaxis=dict(title="Year", tickmode="linear", dtick=1),
    yaxis=dict(title="Credits", tickformat=","),
    yaxis2=dict(
        title="Retired share",
        overlaying="y",
        side="right",
        tickformat=".0%",
        rangemode="tozero",
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="left", x=0),
    margin=dict(l=40, r=40, t=60, b=40),
    height=540,
)

# Self-contained HTML (easy to host anywhere)
fig.write_html("credits_issued_vs_retired.html", include_plotlyjs=True, full_html=True)

print("Wrote: credits_issued_vs_retired.html")
```
